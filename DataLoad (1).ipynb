```python
dataloader = DataLoader(dataset, batch_size=32, shuffle=True)

# O que ele faz:
#  1° Divide em lotes
#  2° Embaralha
#  3° Itera automaticamente
```

```python
class Model(nn.Module):
  def __init__(self):
    super().__init__()
    self.fc1 = nn.Linear(1,16)
    self.relu = nn.ReLU()
    self.fc2 = nn.Linear(16, 1)

  def forward(self, x):
    return self.fc2(self.relu(self.fc1(x)))

model = Model()
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model = model.to(device)
```

```python
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)

epochs = 20

start = time.time()
# treino aqui

for epoch in range(epochs):

    model.train()
    total_loss = 0

    for X_batch, y_batch in dataloader:

        X_batch = X_batch.to(device)
        y_batch = y_batch.to(device)

        y_pred = model(X_batch)
        loss = criterion(y_pred, y_batch)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    if epoch % 10 == 0:
      print(f"Epoch {epoch}, Loss: {total_loss / len(dataloader):.4f}")


end = time.time()

print("Tempo:", end - start)
```

```python
#  === Pratica ===
from torch.utils.data import Dataset, DataLoader
import torch.nn as nn
import torch
import time

# Primeiro você tem que preparar os dados:
class DataSet(Dataset):
  def __init__(self):
    self.X = torch.linspace(0,10, 100).unsqueeze(1)
    self.y = 2 * self.X + 1 + torch.randn_like(self.X) * 2

dataset = DataSet()

class M(nn.Module):
  def __init__(self):
    super().__init__()
    self.layer = nn.Linear(1, 16)
    self.relu =  nn.ReLU()

  def forward(self, x):
    return self.relu(self.layer(x))

model = M()

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)

criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)

epochs = 100

start = time.time()

for epoch in range(epochs):
  model.train()

  x_batch = dataset.X.to(device)
  y_batch = dataset.y.to(device)

  y_pred = model(x_batch)
  loss = criterion(y_pred, y_batch)

  optimizer.zero_grad()
  loss.backward()
  optimizer.step()

  if epoch % 100 == 0:
    print(f"Epoch {epoch}, Loss: {loss.item():.4f}")

end = time.time()
print("Tempo:", end - start)

model.state_dict()
torch.save(model.state_dict(), "modelo.pth")

```

```python
model_loaded = M()
model_loaded.load_state_dict(torch.load("modelo.pth"))
model_loaded.eval()

with torch.no_grad():
  novo_x = torch.tensor([[5.0]])
  pred = model_loaded(novo_x)
  print("Predição:", pred)
```
